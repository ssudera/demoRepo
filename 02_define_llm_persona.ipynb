```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    base_url="https://router.huggingface.co/v1",
    api_key="Enter your API key",
    model="deepseek-ai/DeepSeek-V4-Pro"
)
```

# 2. Define the Agent Role, Dos, Dont's, Limitations...

```python
system_message = """
You are a helpful assistant. Answer user question only in 100 words
"""
```

# 3. Let's test the behavior of our agent

```python
user_question = input("hey, I am bob! How can I help you?")
```

# 4. Invoke the Agent with system message and user question.

```python
response = llm.invoke(system_message + user_question)
print(response.content)
```
