# 1. Setting up LLM Chat connection with Google Gemini

```python
from langchain_google_genai import ChatGoogleGenerativeAI
import os

os.environ["GOOGLE_API_KEY"] = "Enter Your API Key"
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

llm.invoke("hello")
```

# 2. Setting up LLM Chat Connection with Groq API

```python
from langchain_groq import ChatGroq

import os

os.environ["GROQ_API_KEY"] = "Enter your API key"

llm = ChatGroq(model = "openai/gpt-oss-120b")
llm.invoke("hello")
```

# 3. Setting up LLM Chat Connection with OpenAI API

```python
from langchain_openai import ChatOpenAI

import os
os.environ["OPENAI_API_KEY"] = "Enter Your API Key"

llm = ChatOpenAI(model="gpt-5.4-nano")
llm.invoke("hello")
```

# 4. Hugging Face LLM Chat Connections

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    base_url="https://router.huggingface.co/v1",
    api_key="Put your API Key",
    model="meta-llama/Llama-3.1-8B-Instruct:cerebras",
    max_tokens=500
)

response = llm.invoke("What is the capital of France?")
print(response.content)
```
