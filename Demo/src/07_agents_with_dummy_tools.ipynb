# 1. Importing stuff

```python
from langchain.agents import create_agent
from langchain_core.tools import tool
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
load_dotenv()
```

# 2. Defining Tools

```python

#API to get weather details
@tool
def get_weather(city: str) -> str:
    """Get weather for a given city."""
    return f"It's always sunny in {city}!"

@tool
def search(keyword: str) -> str:
    """search the info for the given keyword"""
    return f"Here are the details about the {keyword}!"
    

```

# 3. Definiing LLM

```python
llm = ChatOpenAI(
    base_url="https://router.huggingface.co/v1",
    api_key="Put your own api key",
    model="meta-llama/Llama-3.1-8B-Instruct:cerebras",
    max_tokens=500
)

```

# 4. Creating Agent

```python
agent = create_agent(
    model=llm,   # llm details
    tools=[get_weather, search],       # available tools to bind with agent
    system_prompt="You are a helpful assistant. You reason and action approach" #system_prompt
)
```

# 5. Invoking Agent

```python
result = agent.invoke({
    "messages": [
        {"role": "user", "content": "How is the weather in New Delhi?"}
    ]
})
```

```python
for msg in result["messages"]:
    if msg.__class__.__name__ == "HumanMessage":
        print(f"\n👤 User: {msg.content}")

    elif msg.__class__.__name__ == "AIMessage":
        if msg.tool_calls:
            print(f"\n🤖 AI (tool call): {msg.tool_calls}")
        else:
            print(f"\n🤖 AI: {msg.content}")

    elif msg.__class__.__name__ == "ToolMessage":
        print(f"\n🛠️ Tool ({msg.name}): {msg.content}")
```
