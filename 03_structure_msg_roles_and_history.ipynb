# 1. Initializing the LLM

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    base_url="https://router.huggingface.co/v1",
    api_key="Enter your API key",
    model="meta-llama/Llama-3.1-8B-Instruct:cerebras",
    max_tokens=500
)
```

# 2. Define the Agent Role, Dos, Dont's, Limitations...

```python
messages = []

system_message = """You are a helpful assistant. Answer user question only in 100 words."""
# This could also be done as SystemMessage(content="You are an helpful assistant"
messages.append({'role':'system', 'content': system_message})
```

# 3. Let's test the behavior of our agent

```python
user_question = input("hey, I am bob! How can I help you?")

messages.append({'role':'user' , 'content': user_question})
```

# 4. Invoke the model with system message and user question.

```python
response = llm.invoke(messages)
messages.append({'role':'assistant' , 'content': response.content})

for message in messages:
    print(message['role'] + ":" + message['content'])
```

```python
messages = [
            {'role':'system', 'content':'You are a helpful assistant'},
            {'role':'human', 'content':'What is Deep Learning ?'}
           ]

assistant_response = llm.invoke(messages)
messages.append({'role':'assistant','content':assistant_response.content})
```
