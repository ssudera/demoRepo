```python
# Install dependency (run once in Colab)
!pip install -q langchain langchain-google-genai

import os
from langchain_google_genai import ChatGoogleGenerativeAI

# Directly specify your Gemini API Key
os.environ["GOOGLE_API_KEY"] = "AIzaSyxxxxxxxxxxxxxxxx"

# Initialize the Gemini model
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",  # Or "gemini-1.5-pro"
    temperature=0,
    timeout=30,       # Give up after 30 seconds
    max_retries=2,    # Don't retry indefinitely
)

prompt = "How can I master Generative AI?"

# Invoke the model
response = llm.invoke(prompt)

print(response.content)
```
